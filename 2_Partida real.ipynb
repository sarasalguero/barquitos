## FUNCIONES

```python
# CREAR TABLERO DEL USUARIO Y DE TERMINATOR ORIGINAL
def crear_tablero(tamaño=(10,10)):
    '''
    Esta función está resevada para crear el tablero del usuario.

    Tan solo has de introducir dos parámetros para crear un tablero a tu gusto.
    Si no introduces ningún argumento, el tamaño por defecto será de 10x10.
    Por defecto, las casillas estarán en blanco. Este atributo no se puede modificar.
    '''
    return np.full(tamaño, ".")


# ACTUALIZAR EL TABLERO DE TERMINATOR PARA QUE LO VEA CONNOR
# hay que llamar a esta función después de cada disparo de Connor en el juego para actualizar tablero_terminator_reflejo, así:
    # actualizar_tablero_reflejo(var.tablero_terminator_original, var.tablero_terminator_reflejo)
def actualizar_tablero_reflejo(tablero_original, tablero_reflejo):
    for fila in range(len(tablero_original)): # itera sobre las filas
        for columna in range(len(tablero_original[0])): # itera sobre las columnas
            # Verifica si en la misma posición hay una "X" o un "-"
            if tablero_original[fila][columna] == "X" or tablero_original[fila][columna] == "-":
                # Actualiza la misma posición en el tablero reflejo
                tablero_reflejo[fila][columna] = tablero_original[fila][columna]

# MUESTRA LOS TABLEROS POR PANTALLA
def mostrar_tableros(tablero_connor, tablero_terminator_reflejo, tablero_terminator_original=None):
    print("\nTablero de Connor:")
    for fila in tablero_connor:
        print(" ".join(fila))
    
    print("\nTablero de Terminator:")
    for fila in tablero_terminator_original:
        print(" ".join(fila))

    print("\nTablero reflejo de Terminator:")
    for fila in tablero_terminator_reflejo:
        print(" ".join(fila))



# AÑADIR BARCO RANDOM PARA TERMINATOR O PARA CONNOR, SIRVE PARA AMBOS
def crear_barco_random(eslora, tablero):
    '''
    Esta función sirve para crear barcos aleatorios en nuestro tablero.

    Tiene dos parámetros:
    1. Eslora: Esta es la longitud del barco, es decir, cuántas casillas va a ocupar en el tablero.
    2. Tablero: El tablero donde se coloca el barco.

    El barco se colocará en el tablero de forma aleatoria eligiendo su situación y la orientación en la que crece (N, S, E, O).

    Además, se asegurará de que el barco no excedan los límites del tablero y de que las coordenadas del barco son consecutivas. Una maravilla de función, oiga.
    '''
    barco_random = []  # Esta lista vacía será lo que nos devuelva con las coordenadas rellenas :D

    orientacion = random.choice(["N", "S", "E", "O"]) # Elige la orientación de manera aleatoria

    # Genera una posición inicial válida, para que no se salga del tablero
    fila_random = random.randint(0, 9) # posición 0 y 9, que son 10 posiciones en total
    columna_random = random.randint(0, 9)


    # Se va a asegurar de que ninguna coordenada se salga por los bordes y, al mismo tiempo, de que sean consecutivas entre sí.
    if orientacion == "N":
        # Se asegura de que el barco no se sale del tablero hacia arriba
        while fila_random - eslora + 1 < 0: # se si cumple, es decir, si se sale del tablero:
            fila_random = random.randint(0, 9) # genera una nueva
        for i in range(eslora): # cuando sea false ( es decir, cuando no se salga del tablero) creará las demás coordenadas
            barco_random.append((fila_random - i, columna_random))
    elif orientacion == "S":
        # Se asegura de que el barco no se sale del tablero hacia abajo
        while fila_random + eslora - 1 > 9:
            fila_random = random.randint(0, 9)
        for i in range(eslora):
            barco_random.append((fila_random + i, columna_random))
    elif orientacion == "E":
        # Se asegura de que el barco no se sale del tablero hacia la derecha
        while columna_random + eslora - 1 > 9:
            columna_random = random.randint(0, 9)
        for i in range(eslora):
            barco_random.append((fila_random, columna_random + i))
    elif orientacion == "O":
        # Se asegura de que el barco no se sale del tablero hacia la izquierda
        while columna_random - eslora + 1 < 0:
            columna_random = random.randint(0, 9)
        for i in range(eslora):
            barco_random.append((fila_random, columna_random - i))
    else:
        print("La orientación que has introducido no es válida.") # lo he dejado porque lo tengo desde el principio y creo que lo podría quitar pero me da cosa, so... aquí se queda ;)

    # Verifica que las coordenadas del barco estén vacías en el tablero
    if all(tablero[fila][columna] == "."
        for fila, columna in barco_random):
    # if all: devuelve true si todos los elemtos son true
        return barco_random
    else:
        return crear_barco_random(eslora, tablero) # si no lo están, llama a la propia función de nuevo


# COLOCAR BARCO
# ¿podría poner un argumento con * para meter x barcos al mismo tiempo?
def colocar_barco(barco, tablero):
    '''
    Esta función sirve para colocar tus barcos en el tablero.
    Sirve tanto para barcos custom como para barcos generados de manera aleatoria.
    
    Tiene dos parámetros:
    1. Barco. Introduce aquí la variable del barco que desees añadir.
    2. Tablero. Di en qué tablero lo quieres añadir.
    '''
    for casilla in barco:
        tablero[casilla] = "O"
    return tablero


# COLOCAR BARCOS RANDOM
def colocar_barcos_random(tablero, esloras_posibles):
    '''
    Esta función sirve para colocar tus barcos random en el tablero.

    Tiene dos parámetros:
    1. Tablero: el tablero donde se colocarán los barcos.
    2. Esloras_posibles: es una lista de esloras posibles para los barcos aleatorios, se puede meter como lista en la función o usando una variable tipo lista.

    Los barcos se colocarán en el tablero de forma aleatoria eligiendo su situación y la orientación en la que crece (N, S, E, O).

    Además, se asegurará de que los barcos no excedan los límites del tablero y de que las coordenadas de los barcos sean consecutivas.
    '''

    for eslora in esloras_posibles:
        barco = crear_barco_random(eslora, tablero)
        for casilla in barco:
            tablero[casilla] = "O"

    return tablero




def disparar_connor(tablero_connor, tablero_terminator_original, vidas_connor, vidas_terminator):
    '''
    La finalidad de esta función es atacar a tu contrincante disparando a un punto concreto.
    
    Tiene dos parámetros.
    1. Casilla. Introduce una tupla con la posición de la casilla que desees atacar. Ej: (0,5)
    2. Tablero. Introduce el nombre del tablero al que odias, digo, al que quieras atacar.

    Ya sabes: "Dobby no mata, solo mutila o hiere de gravedad".
    '''
    disparos_terminator = [] #lista para registrar los disparos

    while vidas_connor > 0 and vidas_terminator > 0:
        print("Connor, es el momento de elegir las coordenadas para disparar a Terminator.")
    
        fila_disparo = int(input("Introduce la fila de la coordenada a la que deseas atacar"))
        columna_disparo = int(input("Ahora, introduce la columna de la coordenada a la que deseas atacar"))
        casilla = (fila_disparo, columna_disparo)
        print(f"Tu disparo a la casilla {casilla} está en camino. Suerte, Sarah.")

        # disparos
        if tablero_terminator_original[casilla] == ".":
            print("Agüita. Has fallado. ¡Es el turno del cyborg!")
            tablero_terminator_original[casilla] = "~"
            print(tablero_terminator_original)
            disparar_terminator(tablero_connor, disparos_terminator, vidas_connor)

        elif tablero_terminator_original[casilla] == "O":
            tablero_terminator_original[casilla] = "X"
            print(tablero_terminator_original)
            vidas_terminator -= 1
            print(f"¡¡Tocado!! ¡Bien, Connor! El cyborg está cada vez más débil, solo tiene {vidas_terminator} vidas. ¡Sigue así!")

        # terminar partida
        if vidas_connor == 0:
            print("Terminator ha vencido. Connor, has perdido todas tus vidas.\nSkynet se ha unido a ChatGPT para terminar con todo.\nSayonara, baby.")
            break
        elif vidas_terminator == 0:
            print("Terminator ha perdido todas sus vidas.\nSarah Connor, ¡enhorabuena, has vencido al Cyborg!\nEs hora de decirle: sayonara, baby.")
            break

        

    # return tablero



def disparar_terminator(tablero, disparos_terminator, vidas_connor):
    while True:
        fila = random.randint(0, len(tablero) - 1)
        columna = random.randint(0, len(tablero[0]) - 1)
        casilla = (fila, columna)

        if casilla not in disparos_terminator:
            disparos_terminator.append(casilla)

            if tablero[casilla] == ".":
                print("Terminator ha fallado y ha dado en el agüita, ¡bien! Es tu turno, Connor.")
                tablero[casilla] = "~"
                break
                # aquí debería volver al turno de connor

            else:
                print("¡Maldicón! Terminator ha dado a uno de tus barcos, vuelve a ser tu turno... ¡suerte!")
                tablero[casilla] = "X"
                vidas_connor -= 1  # Reduce una vida al usuario




# MENSAJE DE BIENVENIDA CUANDO SE INICIA EL JUEGO
def mensaje_bienvenida():
    print("¡Te damos la super bienvenida a Hundir la Flota versión: Sarah Connor vs. Terminator!\n")
    print("Para jugar, deberás seguir unas sencillas instrucciones:")
    print("1. Deberás elegir un modo de juego: demo o partida real.")
    print("2. En el modo demo, se jugará una partida rápida con barcos predefinidos.")
    print("3. En el modo partida real, podrás colocar tus propios barcos en el tablero.")
    print("4. Después, alternarás turnos para disparar y hundir los barcos del oponente.")
    print("5. Gana quien hunda todos los barcos del oponente primero.\n")
    print("Diviértete y... ¡¡Corre Sarah Connor, te persigue el Cyborg!!")


```

## VARIABLES

```python
# Barcos
esloras_demo = [2] # solo tiene uno de dos para hacer la demo más rápida
esloras_reales = [1,1,1,1,2,2,2,3,3,4] # tiene TODOS los barcos necesarios
```

## CÓDIGO

```python

# Función principal del juego
def jugar_hundir_la_flota():
    mensaje_bienvenida()  # Muestra un mensaje de bienvenida al inicio del juego

    # Pregunta al usuario qué modo de juego desea iniciar
    modo_juego = input("Elige un modo de juego (demo/partida real): ").lower()

    # Verifica que el modo seleccionado sea válido
    if modo_juego not in ["demo", "partida real"]:
        print("Modo de juego no válido. Elige 'demo' o 'partida real'.")
        return


    # Inicializa los tableros para Connor y Terminator
    tablero_connor = crear_tablero()
    tablero_terminator_original = crear_tablero()
    tablero_terminator_reflejo = crear_tablero()

    # Coloca los barcos en el tablero de Terminator según el modo elegido
    if modo_juego == "demo":
        vidas_connor = 2
        vidas_terminator = 2

        colocar_barcos_random(tablero_terminator_original, esloras_demo)
        colocar_barcos_random(tablero_connor, esloras_demo)

    else:
        vidas_connor = 20
        vidas_terminator = 20
        
        colocar_barcos_random(tablero_terminator_original, esloras_reales)
        colocar_barcos_random(tablero_connor, esloras_reales)


    # Comienza el juego
    while True:
        # Muestra los tableros en cada turno
        mostrar_tableros(tablero_connor, tablero_terminator_reflejo, tablero_terminator_original)
        disparar_connor(tablero_connor, vidas_connor, vidas_terminator, tablero_terminator_original)
        
        if vidas_connor == 0:
            print("¡Has perdido!")
            break
        elif vidas_terminator == 0:
            print("¡Has ganado!")
            break
        break

    while False:
        # Muestra los tableros en cada turno

        mostrar_tableros(tablero_connor, tablero_terminator_reflejo)
        disparar_connor(tablero_connor, vidas_connor, vidas_terminator)
        
        actualizar_tablero_reflejo(tablero_terminator_original, tablero_terminator_reflejo)

        if vidas_connor == 0:
            print("¡Has perdido!")
            break
        elif vidas_terminator == 0:
            print("¡Has ganado!")
            break
        break


# Llama a la función principal para iniciar el juego
jugar_hundir_la_flota()
```
